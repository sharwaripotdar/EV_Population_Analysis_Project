```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

ev_data = pd.read_csv(r'C:\Users\JK\OneDrive\Desktop\PythonProject\Electric_Vehicle_Population_Data.csv')

print(ev_data.head())
print(ev_data.info())

ev_data.isnull()
ev_data.isnull().sum()
ev_data.dropna()
```

```python
## Overall Distribution of EV Type
sns.set_style("whitegrid")
plt.figure(figsize=(6, 6))

## X - axis
xpoints = ev_data['Electric Vehicle Type'].value_counts().sort_index()

## visualising plot
sns.barplot(x = xpoints.index,y = xpoints.values, palette="viridis")

## Labels and Title
plt.title('Overall Distribution of Electric Vehicle Type')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Number of Vehicles Registered')
plt.tight_layout()
plt.show()    
```

```python
## Ev population over time
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

## X - axis
xpoints = ev_data['Model Year'].value_counts().sort_index()

## visualising plot
sns.lineplot(x = xpoints.index,y = xpoints.values, marker='o', color='green')

## Labels and Title
plt.title('EV Population Over Time')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles Registered')
plt.tight_layout()
plt.show()  

```

```python
## geographical distribution
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

## X - axis
xpoints = ev_data['City'].value_counts().head(10)

## visualising plot
sns.barplot(x = xpoints.values, y = xpoints.index, palette="BrBG")

## Labels and Title
plt.title('EV Registrations in Top Cities')
plt.xlabel('City')
plt.ylabel('Number of Vehicles Registered')

plt.tight_layout()
plt.show()
```

```python
## Distribution of Electric Vehicle Ranges
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

## visualising plot
sns.histplot(ev_data['Electric Range'], bins=30, kde=True)

## Labels and Title
plt.title('Distribution of Electric Vehicle Ranges')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Number of Vehicles')
plt.tight_layout()
plt.show() 
```

```python
# calculating the average electric range by model year

## Grouping model year , calculating avg range of Electric range
average_range_by_year = ev_data.groupby('Model Year')['Electric Range'].mean().reset_index()
plt.figure(figsize=(12, 6))

## visualising plot
sns.lineplot(x='Model Year', y='Electric Range', data=average_range_by_year, marker='o', color='#191970')

## Labels and Title
plt.title('Average Electric Range by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Average Electric Range (miles)')
plt.grid(True)
plt.show()
```

```python
# analyzing the popularity of EV manufacturers
# Limiting to top 10 for clarity
ev_make_distribution = ev_data['Make'].value_counts().head(10)  
plt.figure(figsize=(12, 6))

## visualising plot
sns.barplot(x=ev_make_distribution.values, y=ev_make_distribution.index, palette="RdGy")

## Labels and Title
plt.title('Top 10 Popular EV Makes')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Make')
plt.tight_layout()
plt.show()
```

```python
## popular of EV Models
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

## X - axis
xpoints = ev_data['Model'].value_counts().head(5)

## visualising plot
sns.barplot(x = xpoints.values,y = xpoints.index, palette="RdYlGn")

## Labels and Title
plt.title('Top 5 EV Model')
plt.xlabel('Model')
plt.ylabel('Number of Vehicles Registered')
plt.tight_layout()
plt.show()     
```
